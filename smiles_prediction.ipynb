```python
import deepchem as dc
import numpy as np
import pandas as pd
import tensorflow as tf
from deepchem.metrics import Metric
np.random.seed(42)
tf.random.set_seed(42)
```

```python
df = pd.read_csv("data.csv")
df = df[:50000]

df.isnull().sum()

```

```python
df.to_csv("clean_data.csv", index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
fp_featurizer = dc.feat.CircularFingerprint(size=1024,radius=2)
X_fp = fp_featurizer.featurize(df["smiles"].values)
y = df["homolumogap"].values
dataset_fp = dc.data.NumpyDataset(X_fp, y)
```

```python
from deepchem.models import SklearnModel
from deepchem.splits import RandomSplitter

train_test_splitter = RandomSplitter()
dataset_train,dataset_valid,dataset_test = train_test_splitter.train_valid_test_split(dataset_fp,frac_train=0.8,frac_valid=0.1,frac_test=0.1,seed=42)
```

```python
rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)

model_rf = dc.models.SklearnModel(rf, model_dir="rf_fp_model")
model_rf.fit(dataset_train)
```

```python
y_pred_valid = model_rf.predict(dataset_valid)
y_pred_test = model_rf.predict(dataset_test)
```

```python
mse_valid = mean_squared_error(y_pred_valid, dataset_valid.y)
mse_test = mean_squared_error(y_pred_test, dataset_test.y)
r2_valid = dc.metrics.r2_score(dataset_valid.y, y_pred_valid)
r2_test = dc.metrics.r2_score(dataset_test.y, y_pred_test)
print(f"Validation MSE: {mse_valid:.6f}")
print(f"Test MSE:       {mse_test:.6f}")
print(f"Validation R: {r2_valid:.4f}")
print(f"Test R:       {r2_test:.4f}")
```

```python
import joblib
joblib.dump(model_rf, "rf_fp_model.pkl")
joblib.dump(fp_featurizer, "fp_featurizer.pkl")
```

```python
def predict_gap(smiles):
    X_single = fp_featurizer.featurize([smiles])
    ds = dc.data.NumpyDataset(X_single)
    pred = np.ravel(model_rf.predict(ds))[0]
    return float(pred)

print(predict_gap("O=C1C=CC(=O)c2ccccc12"))
```
